```python
#Data Preprocessing

#importing string library
import string

#function to remove punctuations
def remove_punctuation(text):
  punctuationfree=' '
  punctuationfree="".join([i for i in text if i not in string.punctuation])
  return punctuationfree

#adding punctuationfree column to the dataset
data['clean_tweet']=data['tweet'].apply(lambda x:remove_punctuation(x))
```

```python
#viewing dataset
data.head()
```

```python
#to see the punctuations in the string library
string.punctuation
```

```python
#word tokenization function

import nltk
nltk.download('punkt')
def tokenization(text):
  tokens=nltk.word_tokenize(text)
  return tokens
#adding tokenized column to dataset
data['tokenized_tweet']=data['clean_tweet'].apply(lambda x:tokenization(x))
```

```python
#viewing dataset
data.head()
```

```python
#function to remove stopwords
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  output= [i for i in text if i not in stopwords]
  return output
#adding stopwords removed column to dataset
data['remove_stopwords']=data['tokenized_tweet'].apply(lambda x:remove_stopwords(x))
```

```python
#viewing dataset
data.head()
```

```python
#function for lemmatization
from nltk.stem import WordNetLemmatizer # Capitalize 'W' in 'WordNetLemmatizer'
nltk.download('wordnet')
wordnet_lemmatizer=WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
#adding lemmatized column to dataset
data['lemmatized_tweet']=data['remove_stopwords'].apply(lambda x:lemmatizer(x))
```

```python
#viewing dataset
data.head()
```

```python
#assigning lemmatized tokens to corpus list
corpus = [' '.join(tokens) for tokens in data['lemmatized_tweet']]
```

```python
#viewing first 10 sentences/strings in corpus
corpus[0:10]
```

```python
#Bag of words
```

```python
#vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))
#splitting data as x and y
x=cv.fit_transform(corpus)
y=data.label
```

```python
#modelling using random forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
#predicting the model and checking accuracy
from sklearn import metrics
y_pred=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_pred,y_test))
```

```python
#function to input sentiment and test using the model
def find_sentiment(text):
  input_text = cv.transform([text]) # Transform the input text, rename variable to avoid shadowing built-in 'input'
  pred = clf.predict(input_text)[0] # Get the prediction for the input
  if pred == 1:
    return "Negative"
  elif pred == 0:
    return "Positive"
```

```python
import gensim
#preprocessing
data['clean_data']=data['tweet'].apply(lambda x: gensim.utils.simple_preprocess(x))
```

```python
#viewing the clean data after preprocessing
data.head()
```

```python
#training the model
x_train, x_test, y_train, y_test = train_test_split(data['clean_data'], data['label'], test_size=0.2, random_state=42)
```

```python
w2v_model = gensim.models.Word2Vec(x_train,min_count=1)
```

```python
#checking one word embedding for 'iphone'
w2v_model.wv['iphone']
```

```python
#viewing words similar to 'iphone'
w2v_model.wv.most_similar('iphone')
```

```python
words=w2v_model.wv.index_to_key
```

```python
len(words)
```

```python
#initializing variables for vectors
x_train_vec=[]
x_test_vec=[]
```

```python
#iterate through each word in x_train
for sentence in x_train:
  #initialize an empty list to store vectors for current list
  vector=[]
  for word in sentence:
    #check word in our vocabulry
    if word in words:
      vector.append(w2v_model.wv[word])
  x_train_vec.append(vector)
```

```python
#iterate through each word in x_test
for sentence in x_test:
  #initialize an empty list to store vectors for current list
  vector=[]
  for word in sentence:
    #check word in our vocabulry
    if word in words:
      vector.append(w2v_model.wv[word])
  x_test_vec.append(vector)
```

```python
# Initialize empty lists to store average word vectors
x_train_vect_avg = []
x_test_vect_avg = []

#Define a function to calculate average word vectors
def calc_avg_vect(v):
  if v: # Check if the list is not empty
    return np.mean(np.array(v), axis=0) # Convert list of vectors to NumPy array and calculate mean
  else:
    return np.zeros(100,dtype=float)

#Calculate average vectors for X_train_vec
for v in x_train_vec:
  x_train_vect_avg.append(calc_avg_vect(v))

#Calculate average vectors for X_test_vec
for v in x_test_vec:
  x_test_vect_avg.append(calc_avg_vect(v))
```

```python
#modelling
clw=RandomForestClassifier(n_estimators=100,random_state=42)
clw.fit(x_train_vect_avg,y_train)
```

```python
#prediction and accuracy
from sklearn import metrics
y_pred=clw.predict(x_test_vect_avg)
print('Accuracy: ',metrics.accuracy_score(y_test,y_pred))
```

```python
def test_sentiment(text):
  # Transform the input text using the same method as during training
  input_text_vect = [w2v_model.wv[word] for word in word_tokenize(text) if word in w2v_model.wv] # Access word vectors using []
  input_text_avg = calc_avg_vect(input_text_vect)  # Calculate the average word vector
  input_text_avg_reshaped = input_text_avg.reshape(1, -1)  # Reshape to match model input

  pred = clw.predict(input_text_avg_reshaped)[0]  # Get the prediction
  if pred == 1:
    return "Negative"
  elif pred == 0:
    return "Positive"
```

```python
input= 'What amazing service Apple wont even talk question I unless I pay 1995 stupid support'
test_sentiment(input)
```
