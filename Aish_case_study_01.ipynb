```python
##Gender,Married,Dependents,Self_Employed,LoanAmount,Loan_Amount_Term and Credit_History have missing values
#Gender,Married,Dependents and Self_Employed are categorical columns

# handling missing values in gender column

data['Gender'].unique()
```

```python
data['Gender'].value_counts()
```

```python
# mode is Male

data['Gender']=data['Gender'].fillna('Male')
```

```python
data.isna().sum()
```

```python
# now gender has 0 missing values
# handling missing values in Married column

data['Married'].unique()
```

```python
data['Married'].value_counts()
```

```python
# mode is "Yes"

data['Married']=data['Married'].fillna(data['Married'].mode()[0])
```

```python
data.isna().sum()
```

```python
#now Married column has 0 missing values
# handling missing values in Dependants column

data['Dependents'].unique()
```

```python
data['Dependents'].value_counts()
```

```python
# mode is 0
data['Dependents']=data['Dependents'].fillna(data['Dependents'].mode()[0])
```

```python
data.isna().sum()
```

```python
## handling missing values in Self_Employed  column
data['Self_Employed'].unique()
```

```python
data['Self_Employed'].value_counts()
```

```python
#mode is 'No'
data['Self_Employed']=data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
```

```python
data.isna().sum()
```

```python
# handling numerical missing values of  LoanAmount,Loan_Amount_Term and Credit_History
plt.hist(data['LoanAmount'])
plt.title('Distribution of LoanAmount')
plt.xlabel('LoanAmount')
plt.ylabel('Frequency')
```

```python
# it is skewed data
data['LoanAmount']=data['LoanAmount'].fillna(data['LoanAmount'].median())
```

```python
data.isna().sum()
```

```python
plt.hist(data['Loan_Amount_Term'])
plt.title('Distribution of Loan_Amount_Term')
plt.xlabel('Loan_Amount_Term')
plt.ylabel('Frequency')
```

```python
#skewed data.
data['Loan_Amount_Term']=data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].median())
```

```python
data.isna().sum()
```

```python
plt.hist(data['Credit_History'])
plt.title('Distribution of Credit_History')
plt.xlabel('Credit_History')
plt.ylabel('Frequency')
```

```python
#skewed data
data['Credit_History']=data['Credit_History'].fillna(data['Credit_History'].median())
```
