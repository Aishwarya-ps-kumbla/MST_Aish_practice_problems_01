Data Preprocessing steps
1. Removing punctuations
2. Word Tokenization
3. Removing Stopwords
4. Lemmatization



```python
#Data Preprocessing

#importing string library
import string

#function to remove punctuations
def remove_punctuation(text):
  punctuationfree=' '
  punctuationfree="".join([i for i in text if i not in string.punctuation])
  return punctuationfree

#adding punctuationfree column to the dataset
data['clean_review']=data['review'].apply(lambda x:remove_punctuation(x))
```

```python
#viewing dataframe to compare "review" and "clean_review" columns
data.head()
```

```python
#word tokenization function

import nltk
nltk.download('punkt')
def tokenization(text):
  tokens=nltk.word_tokenize(text)
  return tokens
#adding tokenized column to dataset
data['tokenized_review']=data['clean_review'].apply(lambda x:tokenization(x))
```

```python
#viewing dataframe
data.head()
```

```python
#function to remove stopwords
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
def remove_stopwords(text):
  output= [i for i in text if i not in stopwords]
  return output
#adding stopwords removed column to dataset
data['stopwords_free']=data['tokenized_review'].apply(lambda x:remove_stopwords(x))
```

```python
#viewing dataset
data.head()
```

```python
#function for lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
wordnet_lemmatizer=WordNetLemmatizer()
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
#adding lemmatized column to dataframe
data['lemmatized_review']=data['stopwords_free'].apply(lambda x:lemmatizer(x))
```

```python
#viewing dataset
data.head()
```

```python
#assigning lemmatized tokens to corpus list
corpus = [' '.join(tokens) for tokens in data['lemmatized_review']]
```

```python
#viewing first 10 sentences/strings in corpus
corpus[0:10]
```

```python
# Feature Extraction
# #vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(ngram_range=(1,2))

```

```python
#splitting data as x and y
#corpus is assigned to x and sentiment column is assigned to y
x=cv.fit_transform(corpus)
y=data.sentiment
```

```python
#importing train_test_split and splitting data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=42)
```

MODEL BUILDING

1. Bag of Words
2. simple RNN



```python
#Bag of words
#modelling using random forest classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
#predicting the model and checking accuracy
from sklearn import metrics
y_pred=clf.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_pred,y_test))
```

```python
#function to input sentiment and test using the model
def find_sentiment(text):
  input_text = cv.transform([text]) # Transform the input text, rename variable to avoid shadowing built-in 'input'
  pred = clf.predict(input_text)[0] # Get the prediction for the input
  if pred == 1:
    return "POsitive"
  elif pred == 0:
    return "Negative"
```

```python
#negative sentiment from corpus
input= 'Superbly trashy wondrously unpretentious 80 exploitation hooray'
find_sentiment(input)
```

```python
#positive sentiment from corpus
input= ' I waste money buying one'
find_sentiment(input)
```

Model Building using DEEP Learning
1. RNN

```python
#importing libraries
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense
```

```python
#reshapping x_train and X_test
X_train_reshaped = x_train.toarray().reshape(x_train.shape[0], 1, x_train.shape[1])
X_test_reshaped = x_test.toarray().reshape(x_test.shape[0], 1, x_test.shape[1])
```

```python
#modelling
model = Sequential()
model.add(SimpleRNN(units=32, input_shape=(1, x_train.shape[1])))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train_reshaped, y_train, epochs=4, batch_size=500, validation_data=(X_test_reshaped, y_test))
```

```python
#evaluating the model
loss,accuracy = model.evaluate(X_test_reshaped, y_test)
```
